### Crop Day Dataset

```python
# Find columns with missing values, that are also numeric
missing_numeric_cols = cityData.select_dtypes(include=np.number).columns[cityData.select_dtypes(include=np.number).isnull().any()].tolist()


# Now, loop through all the numeric columns with missing values and create a plot for each
for col in missing_numeric_cols:
    plt.figure(figsize=(10, 6))
    
    # Plot KDE
    sns.kdeplot(cityData[col].dropna(), shade=True)
    
    # Add Mean and Median lines
    plt.axvline(cityData[col].mean(), color='red', linestyle='--', label=f"Mean: {cityData[col].mean():.2f}")
    plt.axvline(cityData[col].median(), color='green', linestyle='-', label=f"Median: {cityData[col].median():.2f}")
    
    # Add titles and labels
    plt.title(f'Probability Density Function of {col}')
    plt.xlabel(f'{col} Concentration')
    plt.ylabel('Density')
    plt.legend()
    
    # Show the plot
    plt.show()
```

### Crop Production dataset

```python
# Calculate the 99th percentile to set a reasonable x-axis limit
p99 = cropProd['Production'].quantile(0.99)

# Create a figure for the plot
plt.figure(figsize=(10, 6))

# Plot the KDE for the 'Production' column
# We plot the whole data, but will zoom in with xlim
sns.kdeplot(cropProd['Production'].dropna(), shade=True)

# Add vertical lines for the mean and median
plt.axvline(cropProd['Production'].mean(), color='red', linestyle='--', label=f"Mean: {cropProd['Production'].mean():.2f}")
plt.axvline(cropProd['Production'].median(), color='green', linestyle='-', label=f"Median: {cropProd['Production'].median():.2f}")

# Set the x-axis limit to the 99th percentile for a better view
plt.xlim(0, p99)

# Add titles and labels
plt.title('Probability Density Function of Crop Production (Zoomed to 99th Percentile)')
plt.xlabel('Production')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```
