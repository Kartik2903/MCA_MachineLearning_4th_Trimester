### Task 2: Missing Value Treatment

1. **City AQI Data (`cityData`):**
   - **Pollutants (PM2.5, PM10, etc.):** These are numeric. We will impute missing values using the **median of each city**. This preserves the local characteristics of air quality. For any remaining missing values (where a city has no data for a column), we use the global median.
   - **AQI:** Impute using city-wise median.
   - **AQI_Bucket:** This is categorical. We will fill it based on the standard AQI ranges (0-50: Good, 51-100: Satisfactory, etc.) once the AQI column is imputed.
2. **Crop Yield Data (`cropProd`):**
   - **Production:** We will impute missing values using the **median of that specific Crop and Season**. Production varies wildly between crops (e.g., sugarcane vs. pulses), so a global mean/median would be highly inaccurate.

**Justification:** Median is preferred over mean because air quality and crop production data often contain extreme outliers (as shown in Task 1 describe), making the mean a biased representative of the central tendency.

```python
# Create copies of original data for PDF comparison
cityData_orig = cityData.copy()
cropProd_orig = cropProd.copy()

# Handling cityData pollutants and AQI
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']

print("Before Imputation (cityData):")
print(cityData[pollutants].isnull().sum())

for col in pollutants:
    cityData[col] = cityData.groupby('City')[col].transform(lambda x: x.fillna(x.median()))
    cityData[col] = cityData[col].fillna(cityData[col].median())

# Fill AQI_Bucket based on AQI
def get_aqi_bucket(aqi):
    if aqi <= 50: return 'Good'
    if aqi <= 100: return 'Satisfactory'
    if aqi <= 200: return 'Moderate'
    if aqi <= 300: return 'Poor'
    if aqi <= 400: return 'Very Poor'
    return 'Severe'

cityData['AQI_Bucket'] = cityData['AQI_Bucket'].fillna(cityData['AQI'].apply(get_aqi_bucket))

print("\nAfter Imputation (cityData):")
print(cityData[pollutants].isnull().sum())

# Handling cropProd Production
print("\nBefore Imputation (cropProd):")
print(cropProd['Production'].isnull().sum())

cropProd['Production'] = cropProd.groupby(['Crop', 'Season'])['Production'].transform(lambda x: x.fillna(x.median()))
cropProd['Production'] = cropProd['Production'].fillna(cropProd['Production'].median())

print("After Imputation (cropProd):")
print(cropProd['Production'].isnull().sum())
```

```python
import warnings
warnings.filterwarnings('ignore')

print("Visualizing Probability Density Functions (PDF) Before and After Imputation")

# Plotting for cityData
for col in pollutants:
    plt.figure(figsize=(10, 4))
    sns.kdeplot(cityData_orig[col].dropna(), label='Before (Original Values)', fill=True, color='blue', alpha=0.5)
    sns.kdeplot(cityData[col], label='After (Imputed Values)', fill=True, color='red', alpha=0.3)
    plt.title(f'Distribution Shift: {col}')
    plt.legend()
    plt.show()

# Plotting for cropProd
plt.figure(figsize=(10, 4))
p99 = cropProd_orig['Production'].quantile(0.99) # For better visibility
sns.kdeplot(cropProd_orig['Production'].dropna(), label='Before', fill=True, color='blue', alpha=0.5)
sns.kdeplot(cropProd['Production'], label='After', fill=True, color='red', alpha=0.3)
plt.xlim(0, p99)
plt.title('Distribution Shift: Crop Production (Zoomed to 99th Percentile)')
plt.legend()
plt.show()
```

**Written Concern (Task 2):**
The high volume of missing data (especially >60% for Xylene) introduces a risk of 'imputation bias'. By using the city-wise median, we assume that the air quality at a specific location is consistent enough to represent missing days. However, extreme pollution events that happened on missing days are lost forever, which might make our cleaned data look slightly 'smoother' than reality.

### Task 5: Handling Extreme Values

**Detection Method:** We will use the Interquartile Range (IQR) method to identify outliers and verify them against the 99th percentile.

**Treatment Choice:** Instead of deleting extreme values (which might represent real, albeit rare, severe pollution events), we will **cap** them at the 99th percentile. This reduces their distorting effect on statistical models while retaining the records.

```python
before_stats = cityData['AQI'].describe()

# Calculate Thresholds
Q1 = cityData['AQI'].quantile(0.25)
Q3 = cityData['AQI'].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR
cap_value = cityData['AQI'].quantile(0.99)

outliers_count = (cityData['AQI'] > upper_limit).sum()

# Apply Capping (Winsorization at 99th percentile)
cityData['AQI_Cleaned'] = cityData['AQI'].clip(upper=cap_value)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=cityData['AQI'], color='red')
plt.title('AQI Before Capping')

plt.subplot(1, 2, 2)
sns.boxplot(x=cityData['AQI_Cleaned'], color='green')
plt.title('AQI After Capping (99th Percentile)')

plt.tight_layout()
plt.show()

print(f"Number of outliers detected (IQR): {outliers_count}")
print(f"Capping value (99th percentile): {cap_value}")
print("\nStats comparison:")
print("Before Clean:", before_stats[['mean', 'std', 'max']])
print("After Clean:", cityData['AQI_Cleaned'].describe()[['mean', 'std', 'max']])
```

**Observations (Task 5):**
- The boxplot comparison shows that 'extreme' outliers beyond the 99th percentile were successfully capped. 
- The Standard Deviation (Std) dropped from ~135 to ~116, confirming that the data is now more stable and less prone to being dominated by single-day spikes. This makes our regression models in Task 8 more reliable.

### Task 8: Merging Datasets and Correlation Analysis

**Transformation:** `cityData` ranges from 2015-2020, while `cropProd` primarily ends in 2015 with very sparse data for that year. Because there is almost zero temporal overlap in (State, Year) pairs, a direct join by year results in an empty dataset.

**New Strategy:** We aggregate both datasets at the **State level**, calculating the average AQI (2015-2020) and the average annual Crop Yield for each state. This allows us to investigate the cross-sectional relationship: do states that are generally more polluted also have lower agricultural productivity?

```python
# Aggregate AQI by State
state_aqi = cityData.groupby('State_Name')['AQI_Cleaned'].mean().reset_index()

# Aggregate Crop Production by State
state_crop = cropProd.groupby('State_Name').agg({
    'Production': 'sum',
    'Area': 'sum'
}).reset_index()
state_crop['Yield'] = state_crop['Production'] / state_crop['Area']

# Merge on State Name
merged_df = pd.merge(state_aqi, state_crop, on='State_Name')

print(f"Merged records: {len(merged_df)}")
display(merged_df.head())

# Relationship Exploration
plt.figure(figsize=(10, 8))
sns.heatmap(merged_df[['AQI_Cleaned', 'Production', 'Area', 'Yield']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix: State-wise AQI vs Agricultural Metrics")
plt.show()

# Scatter Plot: AQI vs Yield
plt.figure(figsize=(10, 6))
sns.regplot(data=merged_df, x='AQI_Cleaned', y='Yield', scatter_kws={'alpha':0.6, 's':100}, line_kws={'color':'red'})
plt.title("State-wise Relationship: Average AQI vs. Crop Yield")
plt.xlabel("Average AQI (2015-2020)")
plt.ylabel("Average Crop Yield")
plt.show()
```

**Findings:**
1. **Negative Correlation between AQI and Yield:** The correlation matrix and regression plot show a slight negative relationship. As AQI increases (worse air), the crop yield tends to decline. This might exist because high pollutants like Ozone can directly damage plant tissues and inhibit photosynthesis.
2. **Strong Correlation between Area and Production:** This is a logical relationship where more land generally leads to more output, confirming the basic integrity of our crop dataset.
